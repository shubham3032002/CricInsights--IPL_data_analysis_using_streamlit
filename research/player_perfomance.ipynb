```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
matches=pd.read_csv("../data/matches.csv")
deliveries=pd.read_csv("../data/deliveries.csv")
matches = matches.rename(columns={'id': 'match_id'})
matches=matches.drop(columns=['date','method','umpire1','umpire2','city'])
deliveries=deliveries.drop(columns=['player_dismissed','dismissal_kind','fielder'],axis=1)
data=matches.merge(deliveries,on='match_id',how='left')
```

```python
data.head()
```

```python
batter_data=data[data['batter']=='SC Ganguly']
match_played=batter_data['match_id'].nunique()
total_runs=batter_data['batsman_runs'].sum()
total_balls_faced=batter_data['ball'].count()
average=total_runs/batter_data['is_wicket'].sum() if batter_data['is_wicket'].sum() >0 else 0
strike_rate=(total_runs/total_balls_faced)*100 if total_balls_faced > 0 else 0
print(f"Matches Played: {match_played}")
print(f"Total Runs: {total_runs}")
print(f"Batting Average: {average}")
print(f"Strike Rate: {strike_rate}")
```

```python
season_stats = batter_data.groupby('season').agg(
    total_runs=('batsman_runs', 'sum'),
    matches_played=('match_id', 'nunique'),
    
).reset_index()

print(season_stats)
```

```python
bowler_data = data[data['bowler'] == 'SC Ganguly']

# Bowling performance
total_overs_bowled = bowler_data['over'].nunique()  # Assuming 'over' is represented by a decimal value (e.g., 4.3 for 4 overs and 3 balls)
wickets_taken = bowler_data[bowler_data['is_wicket'] == 1].shape[0]
runs_conceded = bowler_data['total_runs'].sum()
economy_rate = runs_conceded / total_overs_bowled if total_overs_bowled > 0 else 0

print(f"Bowler Total Runs Conceded: {runs_conceded}")
print(f"Bowler Wickets Taken: {wickets_taken}")
print(f"Bowler Economy Rate: {economy_rate}")
```

```python
# Success rate of batting first vs. chasing teams at each venue
batting_first = data[data['result'] == 'runs']
batting_first_stats = batting_first.groupby('venue')['match_id'].nunique().reset_index()
batting_first_stats.rename(columns={'match_id': 'Batting First Wins'}, inplace=True)

chasing = data[data['result'] == 'wickets']
chasing_stats = chasing.groupby('venue')['match_id'].nunique().reset_index()
chasing_stats.rename(columns={'match_id': 'Chasing Wins'}, inplace=True)
```

```python
# Merging all stats into a single DataFrame
venue_analysis = venue_matches.merge(batting_first_stats, on='venue', how='left')
venue_analysis = venue_analysis.merge(chasing_stats, on='venue', how='left')

# Fill NaN values with 0
venue_analysis.fillna(0, inplace=True)

# Calculate success rate
venue_analysis['Batting First Success Rate (%)'] = (
    (venue_analysis['Batting First Wins'] / venue_analysis['Total Matches']) * 100
).round(2)
venue_analysis['Chasing Success Rate (%)'] = (
    (venue_analysis['Chasing Wins'] / venue_analysis['Total Matches']) * 100
).round(2)
```

```python
def score_intervals(data, selected_venue, bins=[0, 50, 100, 150, 200, 250, 300]):
    # Filter data for the selected venue
    venue_data = data[data['venue'] == selected_venue]
    
    # Group by match_id and inning to calculate total runs for each inning
    inning_scores = (
        venue_data.groupby(['match_id', 'inning'])['total_runs']
        .sum()
        .reset_index()
    )
    
    # Create bins and count scores in each interval
    interval_counts = pd.cut(inning_scores['total_runs'], bins=bins, right=False).value_counts().sort_index()
    
    # Convert to DataFrame for easier display
    interval_df = interval_counts.reset_index()
    interval_df.columns = ['Score Range', 'Frequency']
    
    return interval_df
```
