```python
def head_to_head(matches, team1, team2):
    """
    Calculate head-to-head statistics between two teams.

    Parameters:
    matches (DataFrame): DataFrame containing match data with 'team1', 'team2', and 'winner' columns.
    team1 (str): Name of the first team.
    team2 (str): Name of the second team.

    Returns:
    dict: Head-to-head statistics with matches played, wins, and losses for each team.
    """
    # Filter matches where the two teams played against each other
    h2h_matches = matches[
        ((matches['team1'] == team1) & (matches['team2'] == team2)) |
        ((matches['team1'] == team2) & (matches['team2'] == team1))
    ]
    
    # Total matches played
    total_matches = len(h2h_matches)
    
    # Wins for team1
    team1_wins = len(h2h_matches[h2h_matches['winner'] == team1])
    
    # Wins for team2
    team2_wins = len(h2h_matches[h2h_matches['winner'] == team2])
    
    # Losses for each team
    team1_losses = team2_wins
    team2_losses = team1_wins
    
    # Return the results as a dictionary
    return {
        'Teams': f"{team1} vs {team2}",
        'Total Matches': total_matches,
        f'{team1} Wins': team1_wins,
        f'{team1} Losses': team1_losses,
        f'{team2} Wins': team2_wins,
        f'{team2} Losses': team2_losses
    }

# Example usage
team1 = "Mumbai Indians"
team2 = "Chennai Super Kings"

# Call the function with your matches DataFrame
head_to_head_stats = head_to_head(matches, team1, team2)

# Display the results
head_to_head_stats


```

```python
import pandas as pd

def calculate_team_standings(matches_df):
  """
  Calculates the team standings for a given season of IPL matches.

  Args:
    matches_df: DataFrame containing match data with columns:
      - season
      - team1
      - team2
      - winner

  Returns:
    DataFrame with team standings, including:
      - Rank
      - Team
      - Matches Played
      - Wins
      - Losses
  """

  # Filter for the specified season (replace '2024' with desired season)
  season_matches = matches_df[matches_df['season'] == '2014']

  # Count matches played by each team
  team_matches_played = season_matches.groupby('team1')['team1'].count() + \
                        season_matches.groupby('team2')['team2'].count()

  # Count wins for each team
  team_wins = season_matches.groupby('winner')['winner'].count()

  # Create a DataFrame with team standings
  standings_df = pd.DataFrame({
      'Team': team_wins.index,
      'Matches Played': team_matches_played,
      'Wins': team_wins
  })

  # Calculate losses
  standings_df['Losses'] = standings_df['Matches Played'] - standings_df['Wins']

  # Calculate points (adjust as needed based on IPL scoring rules)
  standings_df['Points'] = standings_df['Wins'] * 2  # Example: 2 points per win

  # Sort by points and wins (for tie-breakers) in descending order
  standings_df = standings_df.sort_values(by=['Points', 'Wins'], ascending=False)

  # Assign ranks
  standings_df['Rank'] = range(1, len(standings_df) + 1)

  return standings_df

# Example usage (assuming 'matches' is your DataFrame)
standings_2024 = calculate_team_standings(matches)
standings_2024
```
